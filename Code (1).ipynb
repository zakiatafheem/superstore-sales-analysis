```python
# find outliers
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1
print("IQR :",IQR)
lower = Q1 - 1.5 * IQR
print("Lower Bound :",lower)
upper = Q3 + 1.5 * IQR
print("Upper Bound:",upper)
outliers = df[(df['Sales'] < lower) | (df['Sales'] > upper)]
print("outliers :",outliers)
# Outliers are retained as they represent genuine high-value orders and are important for business insights.
```

```python
# Montly Sales Analysis using line plot
df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()
plt.figure(figsize=(12,4))
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month Year")
plt.ylabel("Sales")
plt.show()
```

```python
# INSIGHT
# Sales show an overall upward trend over time, indicating business growth.

# There are noticeable fluctuations in monthly sales, suggesting seasonal demand patterns and variations in customer purchasing behavior.
```

```python
# Category wise sales
category_sales = df.groupby("Category")['Sales'].sum()
plt.figure()
plt.bar(category_sales.index,category_sales.values)
plt.title("Sales by Category")
plt.xlabel("CATEGORY")
plt.ylabel("SALES")
plt.show()
```

```python
# INSIGHTS
# Technology category generates the highest total sales, indicating strong demand for electronic and tech-related products.

# Furniture category shows moderate sales, suggesting steady but comparatively lower customer demand than Technology.

# Office Supplies contribute consistent sales with smaller order values but frequent purchases.
```

```python
# Category wise sales
category_sales = df.groupby("Sub-Category")['Sales'].sum()
plt.figure(figsize=(12,4))
plt.bar(category_sales.index,category_sales.values)
plt.title("Sales by Category")
plt.xticks(rotation=70)
plt.xlabel("SUB-CATEGORY")
plt.ylabel("SALES")
plt.show()
```

```python
# INSIGHTS
# Phones and Chairs are the highest-selling sub-categories, indicating strong customer demand for technology and furniture products.

# Binders, Storage, and Tables also contribute significantly to overall sales,showing steady demand across office and furniture-related items.

# Fasteners, Envelopes, and Labels have very low sales,suggesting limited demand or low-value transactions.
```

```python
# Profit distribution
plt.figure()
plt.boxplot(df['Profit'])
plt.title("Profit Distribution")
plt.ylabel("Profit")
plt.show()
```

```python
# INSIGHTS

# The profit distribution shows high variability with many outliers, indicating that 
# while some orders generate high profit, several transactions result in losses.
```
